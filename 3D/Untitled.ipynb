```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n = 100

melonwidth = np.random.normal(40, 15, n)
melonlength = np.random.normal(30, 10, n)
melonweight = np.random.normal(9000, 400, n)

balloonswidth = np.random.normal(100, 25, n)
balloonslength = np.random.normal(100, 25, n)
balloonsweight = np.random.normal(100, 10, n)

width = np.concatenate((melonwidth, balloonswidth))
length = np.concatenate((melonlength, balloonslength))
weight = np.concatenate((melonweight, balloonsweight))

X = np.array([width, length, weight]).T

balloonslabel = np.zeros(n, dtype='int')
melonslabel = np.ones(n, dtype='int')

Y = np.concatenate((balloonslabel, melonslabel))
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(X[:n,0], X[:n,1], X[:n,2], c= 'green')
ax.scatter(X[n:,0], X[n:,1], X[n:,2], c= 'red')

ax.set_xlabel('width')
ax.set_ylabel('length')
ax.set_zlabel('weight')
```

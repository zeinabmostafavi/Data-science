```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
tx_data=pd.read_csv('ti_tanic.csv')
tx_data.info()
tx_data.head(10)
```

```python
plt.hist(tx_data.loc[(tx_data['Sex']=='male') & (tx_data['Age']),'Age'], histtype='step', lw=4,
         bins=30, density=True, color='tab:blue', label='male')
plt.hist(tx_data.loc[(tx_data['Sex']=='female') & (tx_data['Age']),'Age'], histtype='step', lw=4,
         bins=30, density=True, color='tab:red', label='female')
plt.title('Age Distribution of Female and Male ')
plt.legend(frameon=False)
plt.xlabel('age')
```

```python
plt.figure(figsize = (16, 8))

plt.hist(tx_data.query("Pclass == 1")['Fare'].to_list(),
         bins=100, density=True, color='tab:red', label='class 1', alpha=0.8)

plt.hist(tx_data.query("Pclass == 2")['Fare'].to_list(),
         bins=100, density=True, color='tab:green', label='class 2', alpha=0.8)

plt.hist(tx_data.query("Pclass == 3")['Fare'].to_list(),
         bins=100, density=True, color='tab:blue', label='class 3', alpha=0.8)

plt.legend(frameon=False)
plt.title('Fare Distribution')
```
